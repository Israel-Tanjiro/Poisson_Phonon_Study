Program to Generate Poisson Distribution For the G4CMP simulations.
This progrma  import the class Poisson_Prob
This class imports a File and calcualte the mean radial probability.
later do a fit to a 5 polynomila grade.
Return the parameter and with the parameter it is possible to obtain the Probability Density for every qubits as a function of the distance

```python
import Poisson_Prob
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
import csv
import math
import numpy as np
from numpy import arange
from scipy.optimize import curve_fit
from matplotlib import pyplot
from scipy.stats import poisson
```

Importing the File with the boundary condition of the simulation

```python
name='Prob_Nb_1_Cu_1_100um_radius.txt'
```

Calling the Class and plotting the Probabily Radial Function and the fit to the data.

```python
Nb=Poisson_Prob.Poisson_Prob(name)
Nb.Ploting_Mean()
PDF=Nb.Fitting_to_Mean()
ax = plt.axes()
polyline = np.linspace(0.0,0.7, 50)
plt.ylabel(' log P(r)  ',fontsize=20)
plt.xlabel('Distance to Qubit Cross [cm]',fontsize=20)

ax.errorbar(Nb.Radius_Not_1, Nb.y_line_log_1_lineal,Nb.Sigma_Nb_1_Cu_1_log_lineal, fmt='.', linewidth=2, capsize=3,label='Data')
plt.plot(polyline, PDF(polyline),'--' ,color='red',label='Fit to Pol 5 ')
plt.plot()
plt.legend(fontsize=15)
plt.show()
```
